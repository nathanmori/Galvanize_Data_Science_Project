```python
def df_line(df, show=True):
    for i, col in enumerate(df.columns):
        if col != 'date':
            df[col].plot()
    if show:
        plt.show()
```

```python
df_line(df_weight)
df_line(df_diet)
```

```python
def avg_7day(ser):
    avg_7day = np.zeros(ser.shape)
    avg_7day.fill(np.nan)
    for i in xrange(7, ser.shape[0]):
        season = ser[i-7:i].reshape(-1,1)
        avg_7day[i] = np.nan if pd.isnull(season).any() else season.mean()
    return avg_7day
```

```python
df_weight_7day = pd.DataFrame(df_input['date'])
df_weight_7day['morning'] = avg_7day(df_weight.morning)
df_weight_7day['night'] = avg_7day(df_weight.night)
```
